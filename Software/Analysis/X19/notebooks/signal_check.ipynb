### check the calibration data

```python
d2,ch2 = getRDF('/data/repos/TMSPlane2/Software/Analysis/X19/scripts/calib_out.root','calib')
```

```python
from ROOT import TGraphErrors, TF1
from array import array

def checkChan(chan): 
    n = ch2.Draw('V:mean:meanErr',f'abs(chan)=={chan}','goff')
    v1 = ch2.GetV1()
    v2 = ch2.GetV2()
    v3 = ch2.GetV3()

    dx = sorted([(v1[i]*7410.,v2[i],v3[i]) for i in range(n)]+[(0,0,0)], key=lambda x:x[0])
#     dx = sorted([(v1[i]*7410.,v2[i],v3[i]) for i in range(n)], key=lambda x:x[0])
    px = lambda t: array('f',[dx[i][t] for i in range(n)])
    gr0 = TGraphErrors(n, px(1), px(0), px(2), array('f',[0]*n))

    #fun1 = TF1('fun1','[0]*x',0,0.2)

    gr0.Draw('APL')
    #gr0.Fit(fun1)
    gPad.Draw()
    
    return gr0
```

```python
from ROOT import TLatex
lt = TLatex()
grx = []
for chan in range(19):
    c = TCanvas('cavT'+str(chan))
    c.cd()
    gr = checkChan(chan)
    gr.Draw('APL')
    #fun1 = gr.GetFunction('fun1')
    #slope = fun1.GetParameter(1)
    lt.DrawLatexNDC(0.2,0.8,f"Ch{chan}")
    c.Draw()
    grx.append(gr)
print(len(grx))
```

```python
from ROOT import TLatex
lt = TLatex()
chan = 10
cx = TCanvas('cav'+str(chan))
cx.cd()
gr = checkChan(3)
gr.Draw('APL')
fun1 = gr.GetFunction('fun1')
slope = fun1.GetParameter(0)
lt.DrawLatexNDC(0.2,0.8,f"Ch{chan} p={slope:.2g}")
cx.Draw()
```
